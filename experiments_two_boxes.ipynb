```python
from dwa_pudla.sym import simulate_two_boxes_process
```

```python
t = 100
a_N = 10
a_S1 = 2
a_S2 = 1
df_total = simulate_two_boxes_process(t, a_N, a_S1, a_S2)
```

```python
df_total.head(20)
```

```python
df_total['time_lasting'] = df_total['time'].shift(-1) - df_total['time']
```

```python
weights = df_total.groupby('in_box_2')['time_lasting'].sum() / t
```

```python
weights = df_total.groupby(df_total['in_box_1'] * df_total['in_box_2'])['time_lasting'].sum() / t
```

```python
weights
```

```python
weights.plot()
```

```python
import numpy as np
np.average(weights.index, weights = weights.values)
```

```python
rolling_num = 1
df_total.set_index('time')['in_box_1'].rolling(rolling_num).mean().plot(title=f'a_N = {a_N}, a_S1 = {a_S1}, a_S2={a_S2}, rolling={rolling_num}', legend='asd')

df_total.set_index('time')['in_box_2'].rolling(rolling_num).mean().plot(title=f'a_N = {a_N}, a_S1 = {a_S1}, a_S2={a_S2} rolling={rolling_num}', legend='asd')
df_total['total'] = df_total['in_box_1'] + df_total['in_box_2']
df_total.set_index('time')['total'].rolling(rolling_num).mean().plot(title=f'a_N = {a_N}, a_S1 = {a_S1}, a_S2={a_S2} rolling={rolling_num}', legend='asd')

```

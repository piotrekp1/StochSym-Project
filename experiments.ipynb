```python
import numpy as np
t = 100000
a_N = 10
a_S = 2
```

```python
N = np.random.poisson(t * a_N)
births = np.sort(np.random.uniform(0, t, N))
live_times = np.random.exponential(1/a_S, N)
deaths = births + live_times
```

```python
import pandas as pd
df = pd.DataFrame({'births': births, 'deaths': deaths, 'live_time': live_times})
df.head()
```

```python
df = pd.DataFrame({'time': births})
df['type'] = 'birth'
df.head()
```

```python
df2 = pd.DataFrame({'time': deaths})
df2['type'] = 'death'
df_total = pd.concat([df, df2])
df_total.head()
```

```python
df_total.sort_values(by='time', inplace=True)
df_total.head()
```

```python
df_total['balance'] = 2 * (df_total['type'] == 'birth') - 1
```

```python
df_total.head()
```

```python
df_total['alive'] = df_total['balance'].cumsum()
```

```python
df_total.head()
```

```python
df_total.set_index('time')['alive'].plot()
```

```python
df_total.set_index('time')['alive'].rolling(10000).mean().plot()
```

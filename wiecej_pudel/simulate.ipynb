```python
from sym import simulate_multibox_from_scheme
import numpy as np
import matplotlib.pyplot as plt
```

```python
t = 1000000
df = simulate_multibox_from_scheme(t, 'kolko')
```

```python
df
```

```python
df['time_lasting'] = df['time'].shift(-1) - df['time']
```

```python
def box_mean(num):
    weights = df.groupby(f'particles_in_{i}')['time_lasting'].sum() / t
    return np.average(weights.index, weights = weights.values)
```

```python
for i in range(1, 5):
    print(f'{i} mean: {box_mean(i)}')
```

```python
for i in range(1, 5):
    axes = df.set_index('time')[f'particles_in_{i}'].plot()
    
    axes.set_ylim([0,df[[f'particles_in_{i}' for i in range(1, 5)]].max().max() + 0.5])
    plt.show()
```
